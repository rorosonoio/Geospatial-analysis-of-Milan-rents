```python
import requests 
import pandas as pd
from bs4 import BeautifulSoup as bs
```

After downloading the needed packages, the data is imported from the Italian renting website immobiliare.it, which lists rent announcements of all italian cities.

```python
r_im = requests.get('https://www.immobiliare.it/mercato-immobiliare/lombardia/milano/') 
soup_im = bs(r_im.content) 

#print('Classes of each table:')
#for table in soup_im.find_all('table'):
#    print(table.get('class'))

tab = soup_im.find('table', class_='nd-table')
immobiliare = pd.read_html(str(tab))[0]
#print(immobiliare)
```

```python
#changing the column names to English and something more easy to use
immobiliare = immobiliare.rename(columns = {'Vendita(€/m²)':'Sale', 'Affitto(€/m²)' : 'Rent'})
```

```python
#the rent is written without the decimals, as italian decimal numbers are written with a comma, instead of the dot.
#This function fixes this issue
def fixrent(number):
    n1 = str(number)
    n2 = list(n1)
    n2.insert(2,'.')
    return float(''.join(n2))

print(fixrent(2345))
```

```python
immobiliare.Rent = immobiliare.Rent.apply(lambda x: fixrent(x)) #previous function is applied
```

```python
#same issue with the sales numbers, now fixed
def fixsale(number):
    number = number*1000
    return int(number)

immobiliare.Sale = immobiliare.Sale.apply(lambda x: fixsale(x))
immobiliare
```
